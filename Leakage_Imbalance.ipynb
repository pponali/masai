# Task 3 — Compute Baseline MAE Build a baseline model that predicts the mean selling_price for every record. Calculate and print the MAE of this baseline on the full cleaned dataset.

```python
df.columns
```

```python

x = df [['car_name', 'brand', 'model', 'vehicle_age', 'km_driven',
       'seller_type', 'fuel_type', 'transmission_type', 'mileage', 'engine',
       'max_power', 'seats']]

y = df['selling_price']

x_temp, x_test, y_temp, y_test = train_test_split(x,y, train_size=.80, random_state=42)

x_train, x_val, y_train, y_val = train_test_split(x_temp, y_temp, test_size=0.18, random_state=42)
```

```python
base_line_prediction = np.mean(y_train)
base_line_predictions = np.full(len(y_val), base_line_prediction)
```

```python
from sklearn.metrics import mean_absolute_error

baseline_mae = mean_absolute_error(y_val, base_line_predictions)
print(f"Baseline MAE : {baseline_mae}")
```
