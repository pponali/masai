# Create a pandas DataFrame using the employee data

```python
import pandas as pd

# dataframe = pd.read_csv("employee.csv")
data = {
    "EmployeeID": [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    "Name": ["Sarah", "Michael", "Jennifer", "David", "Lisa", "James", "Emma", "Robert", "Sophia", "William"],
    "Department": ["Engineering", "Sales", "Marketing", "Engineering", "HR", "Sales", "Engineering", "Marketing", "Sales", "HR"],
    "Salary": [85000, 62000, 71000, 92000, 58000, 68000, 78000, 75000, 65000, 60000],
    "Experience": [4, 3, 5, 5, 4, 4, 3, 4, 5, 3],
    "Rating": [5, 3, 7, 8, 4, 6, 2, 5, 4, 3]
}
dataframe = pd.DataFrame(data)
dataframe
```

# how coloumns with Name, Department and Salary.

```python
dataframe[['Name', 'Department', 'Salary']]
```

# Filter the employees based on the Engineering department.

```python
dataframe[dataframe['Department'] == 'Engineering']
```

# Filter the employees based on the salary greater than 70000.

```python
dataframe[dataframe['Salary'] > 70000]
```

# Filter the employees based on  sales department and performance rating greater than or equal to 4.

```python
dataframe[(dataframe['Department'] == 'Sales') & (dataframe['Performance Score'] >= 4)]
```
