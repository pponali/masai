```python
import sqlite3

conn = sqlite3.connect('bookstore.db')

cursor = conn.cursor()

query = '''
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT NOT NULL,  
    price REAL NOT NULL,
    stock_quantity INTEGER DEFAULT 0
    ); 
    '''

cursor.execute(query)

conn.commit()

conn.close()
```
