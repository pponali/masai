```python
import json

user_profile = {
    "name": "Prakash Ponali",
    "age": 41,
    "email": "prakash.ponali@example.com",
    "is_active": True,
    "skills": ["Python", "Data Analysis", "Machine Learning"]
}

json_string = json.dumps(user_profile, indent=4)
print(json_string)
```

```python
json_string = '{"status": "success", "data": {"user_id": 101, "username": "alex99", "score": 87.5}}'
json_data = json.loads(json_string, object_hook=lambda d: {k: v for k, v in d.items()})
username = json_data["data"]["username"]
score = json_data["data"]["score"]
print(f" The username is {username}")
print(f" The score is {score}")
print(f"User {username} scored {score} points")
```
