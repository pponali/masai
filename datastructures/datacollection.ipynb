```python

#task 2
shopping_list = [
    {"item": "Milk", "price": 50},
    {"item": "Bread", "price": 30},
    {"item": "Eggs", "price": 60},
    {"item": "Rice", "price": 120}
]
shopping_list.append({"item": "Butter", "price": 80})

total_cost = 0
most_expensive_item = 0
most_expensive_item_name = ""

for item in shopping_list:
  if(most_expensive_item < item["price"]):
    most_expensive_item = item["price"]
    most_expensive_item_name = item["item"]
    
  total_cost = total_cost + item["price"]
  

print(f" Total cost of the items : {total_cost}")
print(f" Most expesive Item name : {most_expensive_item_name} and price is : {most_expensive_item}")
    




    
```

```python
# task 3
summary = {}
summary["total_items"] = len(shopping_list)
summary["total_cost"] = total_cost
summary["average_price"] = round(total_cost/len(shopping_list),2)

print(f"Summary : {summary}")
```
